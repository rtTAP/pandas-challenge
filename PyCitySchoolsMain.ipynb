Summary of Analysis
    
    Overall, it appears that the schools budget is not a strong indicator of school performance or student outcomes.
    Students attending Independent schools may have a higher chance of performing better.  
    A schools performance was echoed acorss all year levels within around 2 percent.

Two Conclusions

    Small schools have better average results than medium or larger schools

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python
# File to Load 
school_data = Path("PyCitySchools/Resources/schools_complete.csv")
student_data = Path("PyCitySchools/Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=['school_name', 'school_name'])
#school_data_complete.head(5)
```

# Local Government Area Summary

```python
# Total number of unique schools
total_school = len(school_data)

# Total number of students
total_student = len(student_data)

# Total budget
total_budget = sum(school_data['budget'])

print('Total number of unique schools ' + str(total_school))
print('Total number of students ' + str(total_student))
print('Total budget $' + str(total_budget))
```

```python
# Average maths score
avg_maths_score = student_data['maths_score'].mean()

# Average reading score
avg_reading_score = student_data['reading_score'].mean()

print('Average math score: '+ str(avg_maths_score))
print('Average reading score: '+ str(avg_reading_score))
```

```python
# % passing maths
passing_maths_count = len(school_data_complete[school_data_complete['maths_score'] >= 50])
passing_math_percentage = ( passing_maths_count / total_student )*100

# % passing reading
passing_reading_count = len(school_data_complete[school_data_complete['reading_score'] >= 50])
passing_reading_percentage = ( passing_reading_count / total_student )*100

# Total passing overall
passing_overall_count = len(school_data_complete[(school_data_complete['maths_score'] >= 50) & (school_data_complete['reading_score'] >= 50)])
passing_overall_percentage = (passing_overall_count / total_student)*100

print('Pecentage of passing maths is ' + str(passing_math_percentage) + '%')
print('Pecentage of passing reading is ' + str(passing_reading_percentage) +'%')
print('Pecentage of passing overall is ' + str(passing_overall_percentage) +'%')
```

```python
#create data frame
lga_summary = {
    'Total Unique Schools': [total_school],
    'Total Students': [total_student],
    'Total Budget': [total_budget],
    'Average Maths Score': [avg_maths_score],
    'Average Reading Score': [avg_reading_score],
    'Percentage Passing Maths': [passing_math_percentage],
    'Percentage Passing Reading': [passing_reading_percentage],
    'Overall Passing Percentage': [passing_overall_percentage]   
}

lga_summary = pd.DataFrame(lga_summary)

#formatting
lga_summary['Total Students'] = lga_summary['Total Students'].map("{:,}".format)
lga_summary['Total Budget'] = lga_summary['Total Budget'].map("${:,.2f}".format)

lga_summary
```

# School Summary

```python
# Make school name group data by this
school_name = school_data_complete.set_index('school_name').groupby(['school_name'])
```

```python
# Use the code provided to select the type per school from school_data
school_type = school_data.set_index(['school_name'])['type']
```

```python
# Calculate the total student count per school from school_data
total_students = school_name['Student ID'].count()
total_students
```

```python
# Calculate the total school budget and per capita spending per school from school_data
total_school_budget = school_data.set_index('school_name')['budget']

#budget per student
budget_per_student = budget_per_student = (school_data.set_index('school_name')
                                           ['budget']/school_data.set_index('school_name')['size'])
#total_school_budget
#budget_per_student
```

```python
#calculate the average Maths test scores per school from school_data_complete
average_maths_score = school_name['maths_score'].mean()
#average_maths_score.head(5)
```

```python
#calculate the average Reading test scores per school from school_data_complete
average_reading_score = school_name['reading_score'].mean()
#average_reading_score.head(5)
```

```python
#% Maths Score
pass_maths_percent = school_data_complete[school_data_complete
                                          ['maths_score'] >= 50].groupby('school_name')['Student ID'].count()/total_students*100
#pass_maths_percent.head(5)
```

```python
#% Reading Score
pass_reading_percent = school_data_complete[school_data_complete
                                         ['reading_score'] >= 50].groupby('school_name')['Student ID'].count()/total_students*100
#pass_reading_percent.head(5)
```

```python
#% of students that have passed maths & reading
overall_pass = school_data_complete[(school_data_complete['reading_score'] >= 50) & 
                                    (school_data_complete['maths_score'] >= 50)
                                    ].groupby('school_name')['Student ID'].count()/total_students*100
#overall_pass
```

```python
#create the school summary dataframe
school_summary = pd.DataFrame({
    'School Type': school_type,
    'Total Students': total_students,
    '$ Per Student Budget': budget_per_student,
    'Total School Budget': total_school_budget,
    'Average Maths Score': average_maths_score,
    "Average Reading Score": average_reading_score,
    '% Passing Math': pass_maths_percent,
    '% Passing Reading': pass_reading_percent,
    '% Overall Passing': overall_pass
})

school_summary = pd.DataFrame(school_summary)
school_summary.index.name = 'School Name'
# Formatting
school_summary['Total School Budget'] = school_summary['Total School Budget'].map('${:,.2f}'.format)
school_summary['$ Per Student Budget'] = school_summary['$ Per Student Budget'].map('{:,.2f}'.format)


school_summary
```

# Highest-Performing Schools (by % Overall Passing)

# Scores by School Spending

```python
# Establish the bins
bins = [0, 585, 630, 645, 680]

#bin names
spend_range = ['<$585', '$585-630', '$630-645', '$645-680']
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
#  This step can be skipped but its best to make a copy.
school_spending_df = school_summary.copy()
```

```python
# Change formatting of per student budget in school_summary from string back to float so it can be binned 

school_spending_df["$ Per Student Budget"] = school_spending_df["$ Per Student Budget"].astype(float)
#.astype(float)
```

```python
#cut into the data for desired bin items and range items
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["$ Per Student Budget"], 
                                                         bins, labels=spend_range)
```

```python
#BUG FutureWarning: below code to suppress the message
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
#create the groupings for spend 
spend_summary = school_spending_df.groupby('Spending Ranges (Per Student)')
#develop the spend summary
spend_summary = spend_summary[['Average Maths Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']]

spend_summary.mean()
```

# Scores by School Size

```python
# Establish the bins.
binss = [0, 1000, 2000, 5000]
school_size = ['Small (<1000)', 'Medium (1000-2000)', 'Large (2000-5000)']
```

```python
# Categorize the spending based on the bins
school_spending_df['School Size'] = pd.cut(school_spending_df['Total Students'], 
                                                         binss, labels=school_size)
```

```python
#create the grouping for data
size_summary = school_spending_df.groupby('School Size')
#develop the school size summary
size_summary = size_summary[['Average Maths Score', 'Average Reading Score', '% Passing Math', 
                             '% Passing Reading', '% Overall Passing']]

size_summary.mean()
```
